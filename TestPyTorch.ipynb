```python
class FaceLandmarksDataset():
    
    def __init__(self, csv_file, root_dir):
        """
        Args:
            csv_file (string): Path to the csv file with annotations.
            root_dir (string): Directory with all the images.
            transform (callable, optional): Optional transform to be applied
                on a sample.
        """
        self.entres = csv_file
        self.sorties = root_dir
#         self.transform = True
        
#         self.transform = transforms.Compose(
#              [transforms.ToTensor(),
#              transforms.Normalize(mean=0, std=1)])
        
    
    def __len__(self):
        return len(self.entres)

    def __getitem__(self, idx):
        entre_name = self.entres[idx]
        entre=[]
        with open(entre_name, 'r') as f:
            x=str.split(f.read(), "\n")
            for i in range(len(x)):
                entre.append(x[i])
        for j in range(len(entre)):
            entre[j]=float(entre[j])
        entre=np.array(entre)
#         entre = self.transform(entre)
        entre=torch.tensor(entre)
        entre=entre.double()
        entre=entre.unsqueeze(0)
        entre=entre.unsqueeze(0)
        
        
        sortie_name = self.sorties[idx]
        sortie=[]
        with open(sortie_name, 'r') as f:
            x=str.split(f.read(), "\n")
            for i in range(len(x)):
                sortie.append(x[i])
        for j in range(len(sortie)):
            sortie[j]=float(sortie[j])
        sortie=np.array(sortie)
        sortie=torch.tensor(sortie)
        sortie=sortie.float()
        sortie=sortie.unsqueeze(0)
        
        entre=(entre-(torch.mean(entre)))/torch.std(entre)
#         print("BLABLABLA")
        sample = {'entre': entre, 'sortie': sortie}
        return sample
```
